# filling missing value

```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
df = df.set_index('day')
```

```python
new_df = df
```

```python
new_df = df.fillna(0)
```

```python
new_df
```

```python
new_df = df.fillna({
    'temprature' : 0,
    'windspeed' : 0,
    'event' : 'No event'
})
new_df
```

```python
new_df = df.fillna(method='bfill')
```

```python
new_df
```

```python
new_df = df.fillna(method='ffill')
new_df
```

```python
new_df = df.fillna(method='ffill', limit = 1)#here we can give the limit to fill only one forword fill
new_df
```

```python
new_df = df.fillna(method='ffill', limit = 1, axis = 1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

# dorping the row

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how= 'all') # to delete row which contains all the null
new_df
```

```python
new_df = df.dropna(thresh = 1) # drop non null 1 
new_df
```

```python
new_df = df.dropna(thresh = 2) 
new_df
```

```python
df.dropna(thresh=0.3*len(df.columns))#it will drop those rows which is having only 30% of data
```

```python
df.dropna(thresh=7, axis=1)
```
