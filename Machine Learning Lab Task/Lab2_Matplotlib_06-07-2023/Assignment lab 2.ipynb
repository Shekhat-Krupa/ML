```python
import numpy as np
import matplotlib.pyplot as plt
```

### adjusting the subplots using subplot_adjust()

```python
# plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

#the figure has 1 row, 2 column, 1st chart
plt.subplot(2,2,1)
plt.title("PLOT: 1")
plt.plot(x,y)

#plot 2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

#the figure has 1 row, 2 column, 2nd chart
plt.subplot(2,2,2)
plt.title("PLOT: 2")
plt.plot(x,y)

#plot 3
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,2,3)
plt.title("PLOT: 3")
plt.plot(x,y)

# plot 4
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,2,4)
plt.title("PLOT: 4")
plt.plot(x,y)

plt.suptitle("SUPER TITLE")

#adjusting the sub plots for better view
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
```

### ovelraping barchart 

```python
bars = ['A', 'B', 'C', 'D', 'E', 'F']
part1 = [7, 13, 8, 4, 10, 6]
part2 = [10, 8, 15, 9, 4, 7]
```

```python
x_axis = np.arange(len(bars))
plt.bar(x_axis -0.2, part1, width=0.4, label = 'part1')
plt.bar(x_axis +0.2, part2, width=0.4, label = 'part2')   
plt.xticks(x_axis, bars)
plt.legend()
plt.show()
```

### emoji as marker using marker code & using image 

```python
# using $\U0001F601$ code
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.plot(x,y, marker='$\U0001F601$', markersize = 25)
```

```python
# using image
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
image = plt.imread('emoji.jpg')
```

```python
image_box = OffsetImage(image, zoom=0.15)
```

```python
fig, ax = plt.subplots()
x, y = np.atleast_1d(x, y)
 
# for each value of (x,y), we create
# an annotation
for x0, y0 in zip(x, y):
    ab = AnnotationBbox(image_box, (x0, y0), frameon=False)
    ax.add_artist(ab)
 
ax.plot(x,y, c='green')
plt.show()
```
